```python
# -------------------------------
# Advent of Code Day 1 - Part 1
# -------------------------------

def process_input(file_path):
    a_list = list()
    b_list = list()
    with open(file_path) as f:
        for line in f.read().splitlines():
            a, b = line.split("   ")
            a_list.append(int(a))
            b_list.append(int(b))
    return a_list, b_list

def main(file_path):
    a_list, b_list = process_input(file_path)
    a_list.sort()
    b_list.sort()
    score = sum(abs(b - a) for a, b in zip(a_list, b_list))
    print(score)

if __name__ == "__main__":
    main("/content/drive/MyDrive/Colab Notebooks/AOC 2024/Input/AOC Day 01 Input.txt")

```
