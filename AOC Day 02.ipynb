```python
# -------------------------------
# Advent of Code Day 2 - Part 2
# -------------------------------

# Parsing the Input into a list of reports
def process_input(file_path):
    with open(file_path) as f:
        reports = f.read().splitlines()
        for i in range(len(reports)):
            reports[i] = list(map(int, reports[i].split()))
    return reports

# Function to determine whether report is strictly increasing
def determine_is_valid_increase(report, Dampened = False ):  # Variable to determine if a report is already "Dampened"
  s = report.copy()
  i = 0

  while i+1 < len(s):
    if 1 <= s[i+1]-s[i] <= 3:
      i += 1
    elif not Dampened:
      return (
        determine_is_valid_increase(s[1:], Dampened=True)
        or determine_is_valid_increase(s[:i+1] + s[i+2:], Dampened=True)
      )
    else:
      return False
  return True

# Function to determine whether report is strictly decreasing
def determine_is_valid_decrease(report, Dampened = False):
  s = report.copy()
  i = 0

  while i + 1 < len(s):
      if 1 <= s[i] - s[i + 1] <= 3:
        i += 1
      elif not Dampened:
        return (
          determine_is_valid_decrease(s[1:], Dampened=True)
          or determine_is_valid_decrease(s[:i+1] + s[i+2:], Dampened=True)
        )
      else:
        return False
  return True


# Testing whether the reports are either a valid increase or decrease
def main(file_path):
    reports = process_input(file_path)
    score = 0

    for i, report in enumerate(reports):
        if determine_is_valid_increase(report):
            score += 1
        elif determine_is_valid_decrease(report):
            score += 1
    print(score)

if __name__ == "__main__":
    main("/content/drive/MyDrive/Colab Notebooks/AOC 2024/Input/AOC Day 02 Input.txt")
```
