```python
# -------------------------------
# Advent of Code Day 3 - Part 1
# -------------------------------

# Read the input file
def process_input(file_path):
    with open(file_path) as f:
        return f.read()

def main(file_path):
    memory = process_input(file_path)
    i = 0
    score = 0

    while i < len(memory):
        char = memory[i]
        if char.isdigit() and memory[i-4:i] == "mul(":  # Check if weâ€™re inside a mul(x,y)
            num1 = ""
            num2 = ""
            second_number = False

            # Parse the mul(x,y) expression
            while i < len(memory):
                if char.isdigit():
                    if not second_number:
                        num1 += char
                    else:
                        num2 += char
                elif char == ",":
                    second_number = True
                elif char == ")":
                    score += int(num1) * int(num2)
                    break
                else:
                    break

                i += 1
                if i < len(memory):
                    char = memory[i]
                else:
                    break
        else:
            i += 1

    print(score)

if __name__ == "__main__":
    main("/content/drive/MyDrive/Colab Notebooks/AOC 2024/Input/AOC Day 03 Input.txt")
```

```python
# -------------------------------
# Advent of Code Day 3 - Part 2
# -------------------------------

def process_input(file_path):
    with open(file_path) as f:
        memory = f.read()
    return memory

def main(file_path):
    memory = process_input(file_path)

    i = 0
    score = 0
    active = True

    while i < len(memory):
        # Enable or disable multiplication based on "do"/"don't" commands
        if memory[i:i+3] == "don":
            active = False
        elif memory[i:i+2] == "do":
            active = True

        char = memory[i]
        if char.isdigit() and memory[i-4:i] == "mul(" and active:
            num1 = ""
            num2 = ""
            second_number = False

            while True:
                if char.isdigit():
                    if not second_number:
                        num1 += char
                    else:
                        num2 += char
                elif char == ",":
                    second_number = True
                elif char == ")":
                    score += int(num1) * int(num2)
                    break
                else:
                    break

                i += 1
                if i < len(memory):
                    char = memory[i]
                else:
                    break
        else:
            i += 1

    print(score)

if __name__ == "__main__":
    main("/content/drive/MyDrive/Colab Notebooks/AOC 2024/Input/AOC Day 03 Input.txt")
```
