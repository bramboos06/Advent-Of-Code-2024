```python
# -------------------------------
# Advent of Code Day 7 - Part 1
# -------------------------------

def open_file(file_path):
    with open(file_path) as f:
        f = f.read()
    return f

def process_input(f):
    f = f.splitlines()
    output = []
    for line in f:
        line = line.split(" ")
        line[0] = line[0][:-1]
        line = [int(x) for x in line]
        output.append(line)

    return output

def recursive_combination_equation(equation_nums):
    if len(equation_nums) > 1:
        temp = equation_nums.pop(0)
        product_equation = recursive_combination_equation([x * temp if i == 0 else x for i,x in enumerate(equation_nums)])
        sum_equation = recursive_combination_equation([x + temp if i == 0 else x for i,x in enumerate(equation_nums)])
        return product_equation + sum_equation
    else:
        return equation_nums

def main(file_path):
    calibration_equations = process_input(open_file(file_path))
    total_sum = 0

    for calibration_equation in calibration_equations:
        target_sum = calibration_equation[0]
        equation_nums = calibration_equation[1:]
        if target_sum in recursive_combination_equation(equation_nums):
            total_sum += target_sum

    print(total_sum)
if __name__ == "__main__":
    main("/content/drive/MyDrive/Colab Notebooks/AOC 2024/Input/AOC Day 07 Input.txt")


```

```python
# -------------------------------
# Advent of Code Day 7 - Part 2
# -------------------------------


def open_file(file_path):
    with open(file_path) as f:
        f = f.read()
    return f

def process_input(f):
    f = f.splitlines()
    output = []
    for line in f:
        line = line.split(" ")
        line[0] = line[0][:-1]
        line = [int(x) for x in line]
        output.append(line)

    return output

def recursive_combination_equation(equation_nums):
    if len(equation_nums) > 1:
        temp = equation_nums.pop(0)
        product_equation = recursive_combination_equation([x * temp if i == 0 else x for i,x in enumerate(equation_nums)])
        sum_equation = recursive_combination_equation([x + temp if i == 0 else x for i,x in enumerate(equation_nums)])
        concatenate_equation = recursive_combination_equation([int(str(temp) + str(x)) if i == 0 else x for i,x in enumerate(equation_nums)])
        return product_equation + sum_equation + concatenate_equation
    else:
        return equation_nums

def main(file_path):
    calibration_equations = process_input(open_file(file_path))
    total_sum = 0

    for calibration_equation in calibration_equations:
        target_sum = calibration_equation[0]
        equation_nums = calibration_equation[1:]
        if target_sum in recursive_combination_equation(equation_nums):
            total_sum += target_sum

    print(total_sum)
if __name__ == "__main__":
    main("/content/drive/MyDrive/Colab Notebooks/AOC 2024/Input/AOC Day 07 Input.txt")


```
