```python
# -------------------------------
# Advent of Code Day 9 - Part 1
# -------------------------------

def process_input(file_path):
    with open(file_path) as f:
        f = [[int(x), int(i/2)] if i % 2 == 0 else (int(x)) for i, x in enumerate(f.read())]
    return f

def clean_storage(disk_map):
    i = 1
    while i < len(disk_map):
        if type(disk_map[-1]) == int:
            disk_map.pop()
            continue

        if type(disk_map[i]) == int:
            empty_space = disk_map[i]
            file_space = disk_map[-1][0]
            if empty_space > file_space:
                disk_map[i] -= file_space
                disk_map.insert(i, disk_map.pop())
            else:
                disk_map.pop(i)
                disk_map.insert(i, [empty_space, disk_map[-1][-1]])
                disk_map[-1][0] -= empty_space

        i += 1

    return disk_map


def calculate_checksum(disk_map):
    file_index_num = 0
    filesystem_checksum = 0
    for i, file in enumerate(disk_map):
        for _ in range(file[0]):
            filesystem_checksum += file[1] * file_index_num
            file_index_num += 1
    return filesystem_checksum

def main(file_path):
    disk_map = process_input(file_path)
    disk_map = clean_storage(disk_map)
    print(calculate_checksum(disk_map))

if __name__ == "__main__":
    main("/content/drive/MyDrive/Colab Notebooks/AOC 2024/Input/AOC Day 09 Input.txt")
```

```python
# -------------------------------
# Advent of Code Day 9 - Part 2
# -------------------------------

from time import sleep

def process_input(file_path):
    with open(file_path) as f:
        f = [[int(x), int(i/2)] if i % 2 == 0 else (int(x)) for i, x in enumerate(f.read())]
    return f

def generate_string(disk_map):
    result = ""
    for item in disk_map:
        if isinstance(item, list):
            result += str(item[1]) * item[0]
        else:
            result += "." * item
    return result

def clean_storage(disk_map):
    i = -1
    while i > - (len(disk_map)) - 1:
        if type(disk_map[i]) == int:
            i -= 1
        elif type(disk_map[i]) == list:
            file_space = disk_map[i][0]
            for entry_index, entry in enumerate(disk_map):
                if entry_index < len(disk_map) + i:
                    if type(entry) == int:
                        if file_space < entry:
                            disk_map[entry_index] -= file_space
                            disk_map.insert(entry_index, disk_map[i])
                            disk_map[i] = disk_map[i][0]
                            break
                        elif file_space == entry:
                            disk_map[entry_index], disk_map[i] = disk_map[i], disk_map[entry_index]
                            break
                else:
                    i -= 1
                    break
            else:
                i -= 1

    return disk_map

def calculate_checksum(disk_map):
    file_index_num = 0
    filesystem_checksum = 0
    for file in disk_map:
        if isinstance(file, int):
            file_index_num += file
        else:
            for _ in range(file[0]):
                filesystem_checksum += file[1] * file_index_num
                file_index_num += 1
    return filesystem_checksum


def main(file_path):
    disk_map = process_input(file_path)
    disk_map = clean_storage(disk_map)
    print(calculate_checksum(disk_map))

if __name__ == "__main__":
    main("/content/drive/MyDrive/Colab Notebooks/AOC 2024/Input/AOC Day 09 Input.txt")
```
