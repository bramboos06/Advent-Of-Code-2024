```python
# -------------------------------
# Advent of Code Day 10 - Part 1
# -------------------------------

import numpy as np

def process_input(file_path):
    with open(file_path) as f:
        input = np.array([[int(x) for x in y] for y in f.read().splitlines()])
    return input

def find_trailheads(topo_map):
    trailheads = set()
    for y, row in enumerate(topo_map):
        for x, num in enumerate(row):
            if num == 0:
                trailheads.add((y, x))
    return trailheads

def detect_trails(topo_map, y, x):
    valid_trails = set()
    cur_val = topo_map[y, x]

    directions = [(-1, 0), (0, 1), (1, 0), (0,-1)]
    if topo_map[y, x] == 9:
        return (y, x)
    else:
        for dy, dx in directions:
            ny, nx = y + dy, x + dx
            if 0 <= ny < len(topo_map) and 0 <= nx < len(topo_map[0]):
                if topo_map[ny, nx] - cur_val == 1:
                    temp = detect_trails(topo_map, ny, nx)
                    if isinstance(temp, tuple):
                        valid_trails.add(temp)
                    else:
                        valid_trails.update(temp)
    return valid_trails


def main(file_path):
    topo_map = process_input(file_path)
    trailheads = find_trailheads(topo_map)
    sum_trailhead_scores = 0
    for trailhead in trailheads:
        y, x = trailhead
        trailhead_score = len(detect_trails(topo_map, y, x))
        sum_trailhead_scores += trailhead_score

    print(sum_trailhead_scores)

if __name__ == "__main__":
    main("/content/drive/MyDrive/Colab Notebooks/AOC 2024/Input/AOC Day 10 Input.txt")
```
