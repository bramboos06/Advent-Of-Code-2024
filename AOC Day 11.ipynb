```python
# -------------------------------
# Advent of Code Day 11 - Part 1
# -------------------------------

def process_input(file_path):
    with open(file_path) as f:
        return [int(x) for x in f.read().split(" ")]

def main(file_path):
    stones = process_input(file_path)
    for _ in range(25):
        stones_new = list()
        for stone in stones:
            if stone == 0:
                stones_new.append(1)
            elif len(str(stone)) % 2 == 0:
                middle = len(str(stone)) // 2
                a, b = str(stone)[:middle], str(stone)[middle:]
                stones_new.extend([int(a), int(b)])
            else:
                stones_new.append(stone * 2024)
        stones = stones_new
    print(len(stones))



if __name__ == "__main__":
    main("/content/drive/MyDrive/Colab Notebooks/AOC 2024/Input/AOC Day 11 Input.txt")

```

```python
# -------------------------------
# Advent of Code Day 11 - Part 2
# -------------------------------

import math

def process_input(file_path):
    with open(file_path) as f:
        return [int(x) for x in f.read().split(" ")]

def num_digits(n):
    return int(math.log10(n)) + 1 if n > 0 else 1

def split_even_digits(n):
    digits = num_digits(n)
    half = digits // 2
    divider = 10 ** half
    return n // divider, n % divider

def main(file_path):
    stones = process_input(file_path)
    for _ in range(75):  # now runs 75 times
        print(_)
        stones_new = []
        for stone in stones:
            if stone == 0:
                stones_new.append(1)
            else:
                digits = num_digits(stone)
                if digits % 2 == 0:
                    a, b = split_even_digits(stone)
                    stones_new.extend([a, b])
                else:
                    stones_new.append(stone * 2024)
        stones = stones_new
    print(len(stones))

if __name__ == "__main__":
    main("/content/drive/MyDrive/Colab Notebooks/AOC 2024/Input/AOC Day 11 Input.txt")
```

```python
from inspect import CO_ITERABLE_COROUTINE
# -------------------------------
# Advent of Code Day 11 - Part 2
# -------------------------------

from functools import cache

def process_input(path_file):
    with open(path_file) as f:
        stones = f.read().split(" ")
        stones = list(map(int, stones))
    return stones

@cache
def count_stones(val: int, blinks: int) -> int:
    if blinks == 0:
        return 1
    if val == 0:
        return(count_stones(1, blinks - 1))
    str_val = str(val)
    len_str_val = len(str(val))
    if len_str_val % 2 == 0:
        return count_stones(int(str_val[:len_str_val//2]), blinks - 1) + count_stones(int(str_val[len_str_val//2:]), blinks - 1)
    else:
        return count_stones(val * 2024, blinks - 1)


def main(path_file):
    stones = process_input(path_file)
    print(sum(count_stones(s, 75) for s in stones))


if __name__ == "__main__":
    main("/content/drive/MyDrive/Colab Notebooks/AOC 2024/Input/AOC Day 11 Input.txt")



```
